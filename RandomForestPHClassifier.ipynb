```python
import numpy as np
import pandas as pd
from sklearn import tree

input_file = "ph_v1_days.csv"
df = pd.read_csv(input_file, header = 0)
df.dropna(subset =['Tea(black/green)'], how = 'all',inplace= True)

df.head()
```

```python
df1 = df[['Liquid', 'Lemon water', 'Tea(black/green)', 'PH_avg' ]]
```

```python
df1.head()
```

```python
df1.dropna(subset =['Lemon water'], how = 'all',inplace= True)
print(df1)
```

```python
df1.PH_avg = df1.PH_avg.astype(int)
```

```python
df1['Lemon water'] = df1['Lemon water'].astype(int)
df1['Tea(black/green)'] = df1['Tea(black/green)'].astype(int)

print(df1)
```

```python
from sklearn.ensemble import RandomForestClassifier

features = list(df1.columns[:3])
features

y = df1['PH_avg']
X = df1[features]


clf = RandomForestClassifier(n_estimators=10)
clf = clf.fit(X, y)
```

```python
print (clf.predict([[1000, 100, 300]]))
```
